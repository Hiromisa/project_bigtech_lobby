```python
import pandas as pd
import numpy as np
pd.set_option("max_colwidth", 400)
```

```python
df = pd.read_csv('bigtech_lobby_issue_2011-2021.csv')
```

```python
df = df[['season', 'year','amazon_issue_code', 'amazon_specific_issue', 'facebook_issue_code',
       'facebook_specific_issue', 'google_issue_code', 'google_specific_issue',
       'apple_issue_code', 'apple_specific_issue', 'microsoft_issue_code',
       'microsoft_specific_issue' ]]

# df.head(1)
```

## Amazon ref

```python
sp0 = df['amazon_specific_issue'][1].lstrip("['").rstrip("']")
sp0= pd.Series(sp0)
sp0 = sp0[0].split('\', \'')
sp0 = pd.Series(sp0)

cd0 = df['amazon_issue_code'][1].lstrip("['").rstrip("']")
cd0 = pd.Series(cd0)
cd0 = cd0[0].split('\', \'')
cd0 = pd.Series(cd0)

tech0 = pd.concat([cd0, sp0], axis=1)

tech0 = tech0.rename(columns={0: 'code', 1:'specific_issue'}).set_index('code')

tech0
```

## code PDF parse

```python
import camelot
```

```python
tables =  camelot.read_pdf('LD Web 2014.pdf', flavor='lattice')
```

```python
code = tables[0].df
```

```python
code1 = code[[0,1]]
```

```python
code2 = code[[2,3]]
```

```python
code1 = code1.rename(columns={0:'Code', 1:'Description'}).drop(0)
```

```python
code2 = code2.rename(columns={2:'Code', 3:'Description'}).drop(0)
```

```python
code_full = pd.concat([code1,code2])
```

```python
code_full.to_csv('lobby_code.csv',index=False)
```
